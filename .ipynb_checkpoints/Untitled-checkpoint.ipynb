```python
import pandas as pd

# Load the datasets
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

# See the shape (rows, columns) of each dataset
print(movies.shape)
print(credits.shape)

# Preview the data
movies.head()
```

```python
# Merge movies and credits on 'title'
movies = movies.merge(credits, on='title')
movies.shape
```

```python
movies.head()
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(3)
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
movies.drop_duplicates(inplace=True)
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.iloc[0]
```

```python
# Convert overview into list of words
movies['overview'] = movies['overview'].apply(lambda x: x.split())

# Combine all text columns into one
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Create new dataframe with only useful columns
new_df = movies[['movie_id', 'title', 'tags']]

# Convert list to string
new_df.loc[:, 'tags'] = new_df['tags'].apply(lambda x: " ".join(x))

new_df.head()
```

```python
# Combine all lists into one 'tags' column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Make a clean copy
new_df = movies[['movie_id', 'title', 'tags']].copy()

# Convert list to string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))

# Show sample
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Limit vocab to most frequent 5000 words, ignore English stopwords
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

vectors.shape  # Should print something like (4806, 5000)
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie = movie.lower()
    if movie not in new_df['title'].str.lower().values:
        print("❌ Movie not found. Please check the title.")
        return
    
    idx = new_df[new_df['title'].str.lower() == movie].index[0]
    distances = similarity[idx]
    
    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    print(f"\n🎬 Movies similar to '{new_df.loc[idx, 'title']}':\n")
    for i in movie_list:
        print("👉", new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
